````python
#Ejercicios: Definimos una función círculo el cual permite graficar un círcul de radio R, con centro en c_x, c_y.

import matplotlib.pyplot as plt
import numpy as np



def plot_circulo(radio, c_x, c_y):
  theta = np.linspace(0, 2*np.pi, 100)
  x = c_x + radio*np.cos(theta)
  y = c_y + radio*np.sin(theta)
  plt.plot(x, y)
  

plot_circulo(20, 10,10)
plot_circulo(5, 60,10)
plot_circulo(7, 10,10)


plt.show()
````

````python
## 1). Imprimir un patrón de asteriscos en forma de triángul
def patron_tr(n):
  for i in range(1,n+1):
    for j in range(n-i):
      print(" ", end="")
    for k in range(i):
      print("*"+ " ", end="")
    for l in range(i):
      print(" ", end="")
    print()
patron_tr(8)
````

````python
## 2). Imprimir una tabla de multiplicar del 1 al 10 para todos los números del 1 al 10.
def tab(n):
  for i in range (1,n+1):
    print(f"Tabla del {i}:")
    for j in range (1,n+1):
      print(i , " * " ,  j , " = ", i*j)
    print()
tab(10)
````

````python
## 3). Imprimir un patrón de asteriscos en forma de diamante.
def patron_tr(n):
  for i in range(1,n+1):
    for j in range(n-i):
      print(" ", end="")
    for k in range(i):
      print("* ", end="")
    print()
  for i in range(n,1,-1):
    for j in range(n-i):
      print(" ", end="")
    for k in range(i):
      print("* ", end="")
    print()
patron_tr(8)
````
