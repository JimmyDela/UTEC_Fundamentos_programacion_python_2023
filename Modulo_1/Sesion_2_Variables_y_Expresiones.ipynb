# Variables y expreciones en python
-------------------


| Concepto | Descripción |
| --- | --- |
| Variables | Un nombre que se utiliza para referirse a un valor que puede cambiar durante la ejecución del programa. |
| Asignación | El proceso de asignar un valor a una variable. Se utiliza el operador `=` para asignar un valor a una variable. |
| Tipos de datos | Cada valor en Python tiene un tipo de datos asociado, que define el tipo de operaciones que se pueden realizar con ese valor. Algunos de los tipos de datos más comunes son: números enteros (`int`), números en coma flotante (`float`), cadenas de caracteres (`str`), listas (`list`), tuplas (`tuple`), conjuntos (`set`), y diccionarios (`dict`). |
| Expresiones | Una combinación de valores, variables y operadores que se pueden evaluar para producir otro valor. Por ejemplo, `3 + 4` es una expresión que evalúa a `7`. |
| Operadores aritméticos | Se utilizan para realizar operaciones matemáticas básicas. Algunos de los operadores aritméticos más comunes son: suma (`+`), resta (`-`), multiplicación (`*`), división (`/`), y módulo (`%`). |
| Operadores de comparación | Se utilizan para comparar dos valores y evaluar si son iguales (`==`), distintos (`!=`), mayores (`>`), menores (`<`), mayores o iguales (`>=`), o menores o iguales (`<=`). |
| Operadores lógicos | Se utilizan para combinar expresiones booleanas. Algunos de los operadores lógicos más comunes son: `and`, `or`, y `not`. |
| Concatenación | Se utiliza para unir dos o más cadenas de caracteres en una sola cadena. El operador de concatenación es el signo `+`. |
| Formateo de cadenas | Permite crear cadenas de caracteres más complejas a partir de valores de variables. Se utiliza la función `format()` o la f-string (`f"..."`). |
